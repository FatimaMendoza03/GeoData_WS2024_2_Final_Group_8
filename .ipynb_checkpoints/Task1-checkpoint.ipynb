````python
import requests
import pandas as pd

!pip install chardet
import chardet # this is to detect the encoding

base_url = "https://opendata.dwd.de/climate_environment/CDC/observations_germany/climate/annual/kl/historical/"
file_name = "KL_Jahreswerte_Beschreibung_Stationen.txt"
response = requests.get(base_url + file_name)

# detect the encoding (due to german characters)
result = chardet.detect(response.content)
encoding = result['encoding']

content = response.content.decode(encoding)
utf8_content = content.encode('utf-8')
        
# Save the file with UTF-8 encoding
with open(file_name, "wb") as file:
    file.write(utf8_content)
````

````python
headers = terms = [
    "station_id",
    "from_date",
    "to_date",
    "elevation",
    "latitude",
    "longitude",
    "station_name",
    "federal_state",
    "submission"
]

df = pd.read_fwf(
    file_name,
    skiprows=2,
    names=headers,
    index_col="station_id",
    parse_dates=["from_date", "to_date"]
)
````

````python
df
````

````python
df["from_date"] = pd.to_datetime(df["from_date"], format="%Y%m%d")
df["to_date"] = pd.to_datetime(df["to_date"], format="%Y%m%d")
df['elevation'] = df['elevation'].astype(int)
df['latitude'] = df['latitude'].astype(float)
df['longitude'] = df['longitude'].astype(float)
df["station_name"] = df['station_name'].astype(str)
df["federal_state"] = df['federal_state'].astype(str)
df["submission"] = df['submission'].astype(str).str.replace("Frei", "free")

display(df)
````

````python
idx_bayern    = df["federal_state"] == "Bayern"
idx_active = df["to_date"]   > "2022"
idx_early  = df["from_date"]   < "1950"

df_bayern = df[idx_bayern & idx_active & idx_early]
````

````python
df_bayern
````

### Task 1.2

````python
import geopandas as gpd
````

````python
geo_bayern = gpd.GeoDataFrame(df_bayern, geometry=gpd.points_from_xy(df_bayern.longitude, df_bayern.latitude), crs="EPSG:25832")
````

### Task 1.3

````python
import requests
from bs4 import BeautifulSoup

page = requests.get(base_url)

soup = BeautifulSoup(page.content, "html.parser")
````

````python
a_tag_all = soup.find_all("a")
````

````python
hist_data = []
for element in a_tag_all:
    hist_data.append(element.get("href"))
````

````python
hist_data_bayern = []
for element in hist_data[2:]:
    info = element.split("_")
    
    if int(info[2]) in geo_bayern.index:
        hist_data_bayern.append(element)
````

````python
import wget

for station_data in hist_data_bayern:
    wget.download(base_url + station_data)
````

````python
import zipfile

name_products = []

for file in hist_data_bayern:
    with zipfile.ZipFile(file, "r") as z:
        for name in z.namelist():
            if name.split("_")[0] == "produkt":
                name_products.append(name)
        z.extractall("data")
````

````python
name_products
````

### Task 1.4

````python
df_all_temp = pd.DataFrame()
for product in name_products:
    dftmp = pd.read_csv(
        f"data/{product}",
        sep =";",
        parse_dates=["MESS_DATUM_BEGINN", "MESS_DATUM_ENDE"],
        index_col = "MESS_DATUM_BEGINN",
        na_values = [-999.0]
    )
    s = dftmp["JA_TT"].rename(dftmp["STATIONS_ID"][0]).to_frame()
    df_all_temp = pd.merge(df_all_temp, s, left_index=True, right_index=True, how='outer')
    
df_all_temp.index.names = ["MESS_DATUM_BEGINN"]
df_all_temp.index.rename(name = "year", inplace = True)
df_all_temp.index = df_all_temp.index.year
df_all_temp
````

````python
idx_past = df_all_temp.index >= 1950
idx_present = df_all_temp.index <= 2022

df_year_temp = df_all_temp[idx_past & idx_present]
df_year_temp
````

````python
import seaborn as sns
import matplotlib.pyplot as plt
````

````python
sns.set_style('ticks')
fig1, ax1 = plt.subplots(dpi = 300, figsize = (30,10))

sns.heatmap(df_year_temp.T, cmap='coolwarm', ax = ax1)
fig1.savefig('Bayern_Annual_Temp_Stripes_01.png')
````
